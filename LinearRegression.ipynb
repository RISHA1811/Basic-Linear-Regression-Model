```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
x=np.array([1,2,3,4,5])
y=np.array([30,40,50,55,60])
```

```python
plt.scatter(x,y)
plt.xlabel('Hours Of Studies')
plt.ylabel('Marks Obtained')
plt.title('Marks Prediction according to study hour')
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x=x.reshape(-1,1)
model=LinearRegression()
model.fit(x,y)
```

```python
m=model.coef_[0]
c=model.intercept_
print(m)
print(c)
y=y.reshape(-1,1)
```

```python
y_predict=model.predict(y)
plt.scatter(x,y)
plt.plot(x,y_predict)
plt.xlabel('Hours Of Studies')
plt.ylabel('Marks Obtained')
plt.title('Marks Prediction according to study hour')
plt.show()
```

```python
hours=[[7]]
predict_marks=model.predict(hours)
print(predict_marks)
```
